```python
#@title ðŸ”§ Setup
# This lab needs only NetworkX for graph relationships.

import sys, subprocess                                        # For optional install
def pip_install(pkgs):
    for p in pkgs:
        try: __import__(p.split("==")[0])                     # Try import
        except Exception:
            subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", p])  # Install
pip_install(["networkx"])                                     # Ensure NetworkX is ready

import networkx as nx                                         # Directed graphs for 'isâ€‘a' links
print("âœ… Setup complete for Week 6.")
```

```python
#@title ðŸ•¸ï¸ Semantic network with inheritance (lineâ€‘byâ€‘line)

G = nx.DiGraph()                                              # Create a directed graph
G.add_edges_from([                                            # Add 'isâ€‘a' edges: parent -> child
    ("animal", "bird"),
    ("bird", "penguin"),
    ("animal", "mammal"),
    ("mammal", "human"),
   ("animal", "fish"),  # Added concept
    ("fish", "whale")    # Added concept
])

attributes = {                                                # Local attributes at each node
    "animal": {"can_move": True},
    "bird":   {"has_wings": True, "can_fly": True},
    "penguin":{"can_fly": False},                             # Overrides the inherited 'can_fly'
    "human":  {"uses_language": True},
    "fish": {"lives_in_water": True, "has_gills": True},  # Added properties
    "whale": {"is_carnivore": True}                       # Added property
}

def inherit(node, prop):
    """Return the value of 'prop' for 'node', following parents if needed."""
    if prop in attributes.get(node, {}):                      # If property is defined here
        return attributes[node][prop]                         # Return local value
    for parent in G.predecessors(node):                       # Otherwise, check each parent
        val = inherit(parent, prop)                           # Recursively ask the parent
        if val is not None:                                   # If a parent provides a value
            return val                                        # Return it
    return None                                               # Unknown property

print("Can penguins fly?", inherit("penguin", "can_fly"))     # Expect False (local override)
print("Do humans have wings?", inherit("human", "has_wings")) # Expect True/False? (should be None -> inherited?)
print("Can a fish move?", inherit("fish", "can_move"))
print("Does a whale have gills?", inherit("whale", "has_gills"))
print("Does a whale live in water?", inherit("whale", "lives_in_water"))
print("Is a human a carnivore?", inherit("human", "is_carnivore")) # Expected: None
```

```python
#@title ðŸ“¦ Frames with defaults (lineâ€‘byâ€‘line)

class Frame(dict):                                            # A minimal 'frame' as a dict subclass
    def __init__(self, name, **slots):                        # 'name' is the frame type; slots are attributes
        super().__init__(**slots)                             # Initialize the underlying dict with slots
        self.name = name                                      # Save the type name
    def __repr__(self):                                       # How to display the frame
        return f"<Frame {self.name}: {dict(self)}>"           # Pretty print

course = Frame("AI_Course", code="CSC309", credits=2, assessments=["CA1","CA2","CA3"])  # Example frame instance
print(course)                                                 # Show the frame

# Define the generic 'Assessment' frame template
Assessment_Template = Frame("Assessment",
                            type=None,
                            max_score=100,
                            due_date="YYYY-MM-DD",
                            description="Description of the assessment")

# Instantiate the 'CA2' assessment frame with group plan details
CA2_Assessment = Frame("CA2_Assessment",
                       type="Continuous Assessment",
                       max_score=30, # Assuming CA2 is 30% of total score
                       due_date="2024-04-15",
                       description="Group project on Semantic Networks and Frames")

print("\n--- Assessment Frame Template ---")
print(Assessment_Template)

print("\n--- CA2 Assessment Instance ---")
print(CA2_Assessment)

# TODO(Student): Add at least two more concepts/properties to the network and show inheritance queries.
# TODO(Student): Create an 'Assessment' frame template and instantiate your group's CA2 plan.
```
